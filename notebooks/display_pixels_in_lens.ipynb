```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
input_file = open ('x_y_intensity.json')
json_array = json.load(input_file)
```

```python
structured_data = np.array(json_array)

# Find the minimum and maximum coordinates to determine the size of the grid
min_x, min_y = structured_data[:, 0].min(), structured_data[:, 1].min()
max_x, max_y = structured_data[:, 0].max(), structured_data[:, 1].max()

# Calculate the grid size
grid_width = max_x - min_x + 1
grid_height = max_y - min_y + 1

# Create a 2D array of zeros with the size of the grid
grid = np.zeros((grid_height, grid_width))

# Assign pixel values to their respective locations, adjusting for the minimum offset
for x, y, value in structured_data:
    grid[y - min_y, x - min_x] = value

# Now you can use plt.imshow() to display the grid
plt.imshow(grid, cmap='gray', vmin=0, vmax=np.iinfo(np.uint16).max)
plt.colorbar()  # Optionally add a color bar to show the scale
plt.show()
```
