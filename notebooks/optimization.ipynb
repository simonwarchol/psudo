```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 256x256 numpy array with all zeros
img = np.zeros((256, 256))

# Iterate over the rows and columns of the array
for i in range(32):
    for j in range(32):
        # If the sum of the row and column indices is even, set the value to 1
        if (i + j) % 2 == 0:
            img[i*32:(i+1)*32, j*32:(j+1)*32] = 1

# Display the image
plt.imshow(img, cmap='gray')
plt.show()
```

```python
channel_1 = np.copy(img)
# White out top half of image
channel_1[:channel_1.shape[0]//2, :] = 0
plt.imshow(channel_1, cmap='gray')
```

```python
channel_2 = np.copy(img)
# White out top half of imag
channel_2[:, 0:channel_2.shape[0]//2] = 0
plt.imshow(channel_2, cmap='gray')
```

```python
channel_3 = np.copy(img)
# White out top half of image
channel_3[channel_3.shape[0]//2:channel_3.shape[0], :] = 0
plt.imshow(channel_3, cmap='gray')
```

```python
# psuedocolor channel_1 red [1,0,0] and display the colored image
channel_1_red = np.zeros((img.shape[0], img.shape[1], 3))
channel_1_red[:, :, 0] = channel_1
plt.imshow(channel_1_red)
```

```python
channel_2_blue = np.zeros((img.shape[0], img.shape[1], 3))
channel_2_blue[:, :, 2] = channel_2
plt.imshow(channel_2_blue)
```

```python
# Channel 3 purple
channel_3_purple = np.zeros((img.shape[0], img.shape[1], 3))
channel_3_purple[:, :, 0] = channel_3
channel_3_purple[:, :, 2] = channel_3
plt.imshow(channel_3_purple)
```

```python
combined_image = channel_1_red + channel_2_blue + channel_3_purple
plt.imshow(combined_image)
```

```python
# display channel_1_red, channel_2_blue, and channel_3_purple as subplots in a row
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.imshow(channel_1_red)
ax2.imshow(channel_2_blue)
ax3.imshow(channel_3_purple)
```

```python
# Read file overlap.png into a numpy array
overlap0 = plt.imread('Overlap0.png')
overlap50 = plt.imread('Overlap50.png')
```

```python
from scipy.stats import pearsonr

# Flatten the images into 1D arrays
overlap0_flat = overlap0[:, :, 0].flatten() * 0.5
overlap50_flat = overlap50[:, :, 0].flatten() * 0.5

# Calculate Pearson's correlation coefficient
corr_coef, _ = pearsonr(overlap0_flat, overlap50_flat)

print("Pearson's correlation coefficient:", corr_coef)
```

```python
from skimage import data, filters, measure
pcc, pval = measure.pearson_corr_coeff(overlap0[:, :, 0], overlap50[:, :, 0])
print(f"PCC: {pcc:0.3g}, p-val: {pval:0.3g}")
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

clf = linear_model.LinearRegression()

clf.fit(combined_with_signal[:,0:3], combined_with_signal[:,3:6])
```

```python
y_pred = clf.predict(combined_with_signal[:,0:3])
y_true = combined_with_signal[:,3:6]
```

```python
mse = mean_squared_error(y_true=y_true, y_pred=y_pred, squared=True)
mse
```
